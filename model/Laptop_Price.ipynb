```python
# Find value counts for 'Inches'
inches_counts = data['Inches'].value_counts()

# Identify rare inches (less than 10 data points)
rare_inches = inches_counts[inches_counts < 10].index

# Map each rare inch to its nearest common inch
def map_to_nearest_inch(x):
    if x in rare_inches:
        # Find the nearest common inch value
        common_inches = inches_counts[inches_counts >= 10].index
        nearest = min(common_inches, key=lambda y: abs(y - x))
        return nearest
    else:
        return x

data['Inches'] = data['Inches'].apply(map_to_nearest_inch)
```

```python
import re

def parse_memory(mem_str):
    # Initialize storage types
    hdd = ssd = flash = hybrid = 0

    # Split by '+' to handle multiple drives
    for part in mem_str.split('+'):
        part = part.strip()
        # Find the size and type
        match = re.match(r'(\d+)(TB|GB)\s*(HDD|SSD|Flash Storage|Hybrid)', part)
        if match:
            size, unit, dtype = match.groups()
            size = int(size)
            # Convert TB to GB
            if unit == 'TB':
                size *= 1024
            if dtype == 'HDD':
                hdd += size
            elif dtype == 'SSD':
                ssd += size
            elif dtype == 'Flash Storage':
                flash += size
            elif dtype == 'Hybrid':
                hybrid += size
    return pd.Series([hdd, ssd, flash, hybrid])

data[['HDD', 'SSD', 'Flash_Storage', 'Hybrid']] = data['Memory'].apply(parse_memory)
data = data.drop('Memory', axis=1)
```

```python
!pip install scikit-learn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+' -->' + str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
from sklearn.ensemble import RandomForestRegressor  
rf = RandomForestRegressor()
model_acc(rf)
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = { 'n_estimators':[10, 50, 100],
                'criterion':['squared_error', 'absolute_error', 'poisson']}
grid_obj= GridSearchCV(estimator=rf, param_grid=parameters)
grid_fit = grid_obj.fit(X_train, y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, y_test)
```
